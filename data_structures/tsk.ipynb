```python
def generate_dates(row):
    try:
        fw = row.loc["Date (Week 1) - MM/DD/YYYY format"]
        sw = row.loc["Date (Week 2) - MM/DD/YYYY"]
        rel = row.loc["Release date (DD/MM/YYYY)"]
        end = row.loc["End date"]
        range_ = (end - rel).days
        movie = row.loc["Movie name"]
#         print(fw,sw,range_,movie)
        movie_2_dates[movie]["fw"] = fw
        movie_2_dates[movie]["sw"] = sw
        movie_2_dates[movie]["lt"] = (rel,end)
    except Exception as e:
        print(movie,e)
```
